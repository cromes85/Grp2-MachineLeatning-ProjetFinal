```python
import os
colonnes = ["marque","model","year","price","transmission","mileage","fuelType","tax","mpg","engineSize"]
data = pd.DataFrame([],columns=colonnes)
debut = 0
for file in os.listdir("dataset_auto"):
    if '.csv' in file:
        data = pd.concat([data,pd.read_csv("dataset_auto/"+file)], ignore_index=True)
        data.iloc[debut:data.shape[0],0] = file.split('.csv')[0]
        debut = data.shape[0]
data
```

## Fréquence de chaque valeurs de chaque caractérisques 

```python
def affichage_freq():
    for caracteristique in colonnes:
        print("=====",caracteristique,"=====")
        print(data[caracteristique].value_counts())
        print("***========================================***")
affichage_freq()
```

## définition de la fonction effacement

```python
def effacement(data, col, val):
    data.drop(data.loc[data[col]== val].index, inplace=True)
    data.reset_index(drop=True, inplace=True)
    data
```

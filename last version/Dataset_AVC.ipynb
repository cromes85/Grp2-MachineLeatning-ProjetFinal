```python
def getXy(standard =True):      
    X = data_AVC.drop(columns=["stroke"]).values.reshape(-1,16)
    y = data_AVC["stroke"].values
    return X,y
```

```python
from sklearn.model_selection import train_test_split 
def modifieXyTrainTest(X,y):
    Xtrain,Xtest, ytrain,ytest = train_test_split(X,y, random_state = 2, test_size = 0.20)
    return Xtrain, Xtest, ytrain,ytest
```

```python
Xtrain, Xtest, ytrain,ytest = modifieXyTrainTest(X,y)
```

```python
print(Xtrain,ytrain)
print("=================================")
print(Xtest,ytest)
```

```python
from imblearn.over_sampling import RandomOverSampler

ros = RandomOverSampler(random_state=2)
Xtrain, ytrain = ros.fit_resample(Xtrain, ytrain)
from collections import Counter


```

```python
Counter(ytrain)

```

```python
from sklearn.ensemble import IsolationForest

modelIsolationForest = IsolationForest(contamination = 0.01, random_state=2, n_estimators=100)
modelIsolationForest.fit(Xtrain)

plt.scatter(Xtrain[:,3],Xtrain[:,4], c = modelIsolationForest.fit_predict(Xtrain))
# axes= plt.gca()
# print(axes.set_xlim(0,4))
#print("Val axes : ",plt.axis())
plt.axis()
```

```python
outliers = modelIsolationForest.predict(Xtrain)
Xtrain = Xtrain[outliers == 1]
ytrain = ytrain[outliers == 1]
#TGCM
```

```python
sns.pairplot(data_AVC)
plt.show()
```

```python
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier

dtree = DecisionTreeClassifier(random_state=2)
dtree = dtree.fit(Xtrain,ytrain)
```

```python
pred1=dtree.predict(Xtest)
```

```python
pred1
```

```python
dtree.score(Xtest,ytest)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
print(accuracy_score(ytest,pred1))
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(ytest, pred1)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay

cm = confusion_matrix(ytest, pred1)
cm_display = ConfusionMatrixDisplay(cm).plot()

cm_display
```
